##### pattern matching using regex

If you want to replace many words from a list ,
You can build the regex dynamically:

```python
import re
```

```python
text = "hi there, bye now, hii too"
result = re.sub(r"(hi|bye)", "ok", text)
print(result)
# Output: "ok there, ok now, hii too"
```

```python
words = ["hi", "bye", "hello", "ciao"]
pattern = r"|".join(map(re.escape, words))  # escape handles special chars
text = "hi and bye, hello and ciao"
result = re.sub(pattern, "ok", text)
print(result)
# "ok and ok, ok and ok"
```

Loop through a list/tuple of targets

```python
text = "hi there, hii again"
for target in ("hi", "hii"):
    text = text.replace(target, "hello")
print(text)
# "hello there, hello again"
```

```python
import re

text = "hi there, hii again"
text = re.sub(r"hi{1,2}", "hello", text)  # matches "hi" or "hii"
print(text)
# "hello there, hello again"
```
